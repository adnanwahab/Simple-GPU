```python
def decode(message_file):
    with open(message_file, 'r') as file:
        lines = file.readlines()

    # Parsing the file content
    word_dict = {}
    for line in lines:
        number, word = line.split()
        word_dict[int(number)] = word

    # Constructing the pyramid and identifying key numbers
    key_numbers = set()
    current_row = 1
    counter = 1
    while counter in word_dict:
        if counter == current_row * (current_row + 1) // 2:
            key_numbers.add(counter)
            current_row += 1
        counter += 1

    # Forming the decoded message
    decoded_message = ' '.join(word_dict[num] for num in sorted(key_numbers))
    return decoded_message

# Example usage
message = decode('/Users/shelbernstein/Simple-GPU/lib/coding_qual_input.txt')
print(message)
```
